## Pitch mean and std_deviation extraction for Healthy Control

```python
import os
import librosa
import numpy as np
import pandas as pd
```

```python
# Specify the path to the directory containing WAV files
source_path = 'D:/BIT MESRA/Semester 7/Minor Project/Training Data/train/HC/'

source_path;
```

```python
# Get a list of all mp3 files in the directory
HC_train_files = [f for f in os.listdir(source_path) if f.endswith('.mp3')]

HC_train_files;
```

```python
# Initialize lists to store pitch values, mean, and standard deviation
pitch_values = []
mean_pitch = []
std_pitch = []

# Loop through each WAV file
for file_name in HC_train_files:
    file_path = os.path.join(source_path, file_name)
    
    # Load the audio file
    y, sr = librosa.load(file_path)
    
    # Calculate pitch (fundamental frequency)
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    f0 = []
    for i in range(pitches.shape[1]):
        f0.append(pitches[:, i][magnitudes[:, i].argmax()])
    
    # Append pitch values to the pitch_values list
    pitch_values.append(f0)
    
    # Calculate mean pitch value and standard deviation
    M = np.mean(f0)
    mean_pitch.append(M)
    std_pitch.append(np.std(f0))

# Create a DataFrame to store the results
df = pd.DataFrame({'File Name': HC_train_files, 'Mean Pitch': mean_pitch, 'Std Pitch': std_pitch})

# Save the DataFrame to a CSV file
df.to_csv('HC_mean_std_pitch.csv', index=False)
```
