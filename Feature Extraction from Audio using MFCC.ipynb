## Function to extract the 13-mfcc for an audio file

```python
def feature_extraction(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    mfcc = np.mean(librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=13).T, axis=0)
    
    return mfcc
```

```python
features = {}
directory = 'D:/BIT MESRA/Semester 7/Minor Project/Training Data/train/AD/'
for audio in os.listdir(directory):
    audio_path = directory+audio
    features[audio_path] = feature_extraction(audio_path)
```

## Labeling the alzheimers patients data

```python
label = 'Alzheimer'

for audio in os.listdir(directory):
    audio_path = os.path.join(directory, audio)
    features[audio_path] = {'mfcc': features[audio_path], 'label': label}
# Now, 'features' contains both MFCC features and labels for each audio file
```

## Labeling the control data

```python
label = 'Control'

control_features = {}
directory = 'D:/BIT MESRA/Semester 7/Minor Project/Training Data/train/HC/'
for audio in os.listdir(directory):
    audio_path = directory+audio
    control_features[audio_path] = {'mfcc': feature_extraction(audio_path), 'label':label}
```

### Make the keys just the filenames instead of whole path to make the data more readable

```python
control_features_with_filenames = {}
for file_path, data in control_features.items():
    filename = os.path.basename(file_path)
    control_features_with_filenames[filename] = data
```

```python
alzheimers_features_with_filenames = {}
for file_path, data in features.items():
    filename = os.path.basename(file_path)
    alzheimers_features_with_filenames[filename] = data
```

```python
alzheimers_features_with_filenames;
control_features_with_filenames;
```

```python
combined_dataset = {}

# Add Alzheimer's samples to the combined dataset
for filename, data in alzheimers_features_with_filenames.items():
    combined_dataset[filename] = data

# Add Control samples to the combined dataset
for filename, data in control_features_with_filenames.items():
    combined_dataset[filename] = data
```

```python
combined_dataset;
```

## Extract the Combined Labeled Data onto a csv file

```python
import csv

# Define the path for the CSV file
csv_file_path = 'combined_dataset.csv'

# Write the combined dataset to the CSV file
with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = ['Audio_File', 'MFCC_Features', 'Label']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header
    writer.writeheader()
    
    # For each row of combined_dataset, write the row into the csv file
    for filename, data in combined_dataset.items():
        writer.writerow({
            'Audio_File': filename,
            'MFCC_Features': data['mfcc'],
            'Label': data['label']
        })

print(f"Combined dataset exported to {csv_file_path}")
```
