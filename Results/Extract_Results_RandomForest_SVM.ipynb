```python
import pandas as pd
from tabulate import tabulate

# Define the data
data = {
    'Train-Test Split': ['70-30'],
    'Param Grid': ["{'n_estimators': [100,150,200,250,300,350,400,500,700], 'max_features': ['auto', 'sqrt', 'log2'], 'max_depth' : [3,4,5,6,7,8,9,10], 'criterion' :['entropy', 'gini']}"],
    'Cross-Validation': ['LeaveOneOut'],
    'Scoring': ['accuracy'],
    'Grid Search': ['GridSearchCV'],
    'Best Training Accuracy': [0.656433],
    'Best Parameters': [
        {'criterion': 'entropy', 'max_depth': 10, 'max_features': 'sqrt', 'n_estimators': 100}
    ],
    'Testing Accuracy': [0.5],
    'Testing Sensitivity (Abnormality)': [0.6538461538461539],
    'Testing Specificity (Normality)': [0.3181818181818182],
    'Confusion Matrix': [[[7, 15], [9, 17]]],
    'MCC Score': [-0.02956561979945413]
}

# Create a DataFrame with the specified data
df = pd.DataFrame(data)

# Extract columns after 'Grid Search'
columns_after_grid_search = df.columns[df.columns.get_loc('Grid Search')+1:]

# Create a new DataFrame with only the columns after 'Grid Search'
df_after_grid_search = df[columns_after_grid_search]

# Display the DataFrame in a prettier table format
print(tabulate(df_after_grid_search, headers='keys', tablefmt='pretty'))
df_after_grid_search.to_csv('results_Random_Forest.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the data
data = {
    'Metric': ['Best Training Accuracy', 'Testing Accuracy', 'Testing Sensitivity (Abnormality)', 'Testing Specificity (Normality)', 'Confusion Matrix', 'MCC Score'],
    'Value': [0.710784, 0.7, 0.75, 0.65625, '[[21, 11], [7, 21]]', 0.40625]
}

# Create a DataFrame with the specified data
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Hide the axes
ax.axis('off')

# Create a table within the plot
tab = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Style only the headers to make them bold
for key, cell in tab._cells.items():
    if key[0] == 0:  # Check if it's a header cell (first row)
        cell.set_text_props(fontweight='bold')

# Style the table
tab.auto_set_font_size(False)
tab.set_fontsize(12)
tab.scale(1.5, 1.5)

# Save as PNG image
plt.savefig('SVM_Results_default.png', bbox_inches='tight', pad_inches=0.1, dpi=300)

# Save as CSV file
df.to_csv('SVM_Results_default.csv', index=False)

# Show the table within the plot
plt.show()
```
